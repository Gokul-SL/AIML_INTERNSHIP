```python
import pandas as pd
data={
    'Name':['Bob','John','David','Mark','Ryan','Grace','Steve','Peter','Tony','Bruce','Mike','Walter','Harry','Neslon','Wilson','Jack','Carl','Eve','Hank','Frank'],
    'Age':[23,21,28,24,25,24,27,22,29,25,24,28,23,21,26,24,21,22,28,25],
    'Department':['HR','Sales','IT','IT','Finance','IT','Sales','Finance','IT','Sales','HR','HR','Finance','HR','Sales','IT','HR','IT','Finance','IT'],
    'Salary':[30000,35000,50000,45000,42000,50000,45000,50000,55000,45000,35000,35000,45000,40000,45000,45000,40000,50000,30000,45000],
    'Experience':[2,1,5,4,4,3,5,3,6,4,3,3,4,4,5,5,3,6,2,4],
    'Performance_Score':[75,70,95,80,85,90,85,90,100,85,75,75,85,80,85,85,80,90,70,85]
}
df=pd.DataFrame(data)
print(df)
```

```python
#part 1
import numpy as np
Salary_arr=df['Salary'].to_numpy()
Exp_arr=df['Experience'].to_numpy()
print(f"{Salary_arr}\n{Exp_arr}")

mean=np.mean(Salary_arr)
median=np.median(Salary_arr)
sd=np.std(Salary_arr)
print(f"Mean:{mean} \nMedian:{median} \nStandard Deviation:{sd:.2f}")

Max=np.max(Exp_arr)
Min=np.min(Exp_arr)
print(f"Maximum Experience:{Max} \nMinimum Experience:{Min}")

norm_sal=(Salary_arr-np.min(Salary_arr))/(np.max(Salary_arr)-np.min(Salary_arr))
print(f"Normalize Salary:{np.round(norm_sal,2)}")
```

```python
#part 2
import pandas as pd
df.index=range(1,len(df)+1)
print(df.head(5))

print(f"\n{df.describe()}")

avg_sal=df.groupby('Department') ['Salary'].mean()
print(f"\n{avg_sal}")

high_perf=df[df['Performance_Score']==df['Performance_Score'].max()]
print(f"\nTop performer:{high_perf['Name'].values[0]}")

df['Salary_After_Bonus']=df['Salary']*1.10
```

```python
#part 3
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
avg_sal.plot(kind='bar',color='springgreen')
plt.title('Average Salary By Department')
plt.xlabel('Department')
plt.ylabel('Salary')
plt.show()

plt.figure(figsize=(8,5))
sorted_df=df.sort_values('Age')
plt.plot(sorted_df['Age'],sorted_df['Salary'],marker='o',color='crimson',label='Salary Trend')
plt.title('Age vs Salary')
plt.xlabel('Age')
plt.ylabel('Salary')

plt.figure(figsize=(8,8))
df['Department'].value_counts().plot(kind='pie',autopct='%1.1f%%')
plt.title('Department Distribution')
plt.xlabel('')
plt.ylabel('')
plt.show()

plt.figure(figsize=(8,9))
plt.hist(df['Salary'],bins=5,color='hotpink',edgecolor='black')
plt.title('Salary Distribution')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()
```
