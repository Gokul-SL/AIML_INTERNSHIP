```python
import streamlit as st
import joblib
import numpy as np

# Load the saved model
model = joblib.load('fraud_model.joblib')

st.title("Credit Card Fraud Detector")
st.write("Enter transaction details to check for fraud.")


inputs = []
cols = st.columns(2) 

for i in range(10):
    with cols[i % 2]:
        val = st.number_input(f"Transaction Feature {i+1}", value=0.0)
        inputs.append(val)

if st.button("Check Transaction"):
    prediction = model.predict([inputs])
    
    if prediction[0] == 1:
        st.error(" WARNING: This transaction looks FRAUDULENT!")
    else:
        st.success(" Clear: This transaction appears safe.")
```
