```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [10, 14, 12, 18, 20]
plt.plot(x, y, marker='o')
plt.title("Simple Line Chart")
plt.xlabel("Time")
plt.ylabel("Distance")
plt.grid(True)
plt.show()
```

```python
students = ["A", "B", "C", "D", "E"]
marks = [85, 70, 90, 65, 95]
plt.bar(students, marks)
plt.title("Student Marks")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

```python
students = ["A", "B", "C", "D", "E"]
marks = [85, 70, 90, 65, 95]
plt.bar(students, marks, color=["red", "blue", "green", "orange", "purple"])
plt.title("Student Marks")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

```python
activities = ["Study", "Exercise", "Sleep", "Hobbies"]
hours = [4, 1, 7, 2]
df=pd.DataFrame({'Hours':hours},index=activities)
df.plot(kind='pie',y='Hours',figsize=(8,8),autopct='%1.1f%%')
plt.title('Student Activities')
plt.ylabel('')
plt.show()
```
